# Dimensionality Reduction with t-Distributed Stochastic Neighbor Embedding (t-SNE)

```python
# Import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Import additional dependencies
from sklearn.manifold import TSNE

# Ignore Python warnings about past model versions
import warnings
warnings.filterwarnings("ignore")
```

## 3D Visualization of Dataset using t-SNE

```python
# Import datasets
X_train = pd.read_csv("data/x_train.csv", index_col = 0)
X_test = pd.read_csv("data/x_test.csv", index_col = 0)
y_train = pd.read_csv("data/y_train.csv", index_col = 0)
y_test = pd.read_csv("data/y_test.csv", index_col = 0)
```

```python
# Create t-SNE model with 3-dimensional representation
tsne = TSNE(n_components = 3)

# Fit the t-SNE model on the training feature dataset
X_train_tsne_3d = tsne.fit_transform(X_train)
```

```python
# Display the 3D t-SNE visualization
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection = "3d")
ax.scatter(X_train_tsne_3d[:,0], X_train_tsne_3d[:,1], 
           X_train_tsne_3d[:,2])
ax.set_xlabel("t-SNE Component 1")
ax.set_ylabel("t-SNE Component 2")
ax.set_zlabel("t-SNE Component 3")
plt.title("3D t-SNE Visualization of Reduced Spotify Dataset", fontweight = "bold")
plt.show();
```

The 3D rendering of the Spotify dataset with reduced dimensions didn't provide the information or visualizations I had hoped it would. Because of this I decided to scrap my pursuit of further visualization using 3D t-SNE and instead moved to focus on exploratory data analysis on the dataset using a 2D implementation of t-SNE.

## Exploratory Dataset Analysis using 2D t-SNE

```python
# Create a t-SNE model with 2-dimensional representation
tsne = TSNE(n_components = 2)

# Fit the t-SNE model on the training feature dataset
X_train_tsne_2d = tsne.fit_transform(X_train)
```

```python
# Display the visualization for the t-SNE model
plt.figure(figsize = (10,10))
plt.scatter(X_train_tsne_2d[:,0], X_train_tsne_2d[:,1])
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.title("2D t-SNE Visualization of Reduced Spotify Dataset", fontweight = "bold")
plt.show();
```

t-SNE is intended to reveal **local structure** and **relationships** from the various songs throughout the high-dimensional Spotify dataset. The proximity and similarity relationships between these data points actually does reveal some substantial information on the surface level. It can be seen that in terms of local structure, there appears to be a **conglomeration** of data towards the **center** of the **dimension-reduced space**. Additionally this tells us the relationship of a lot of the data by showing there being a highly centralized area of data mass with the rest distributed throughout the space.
